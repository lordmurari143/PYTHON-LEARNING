```python
name=input("enter the name of student =")
english = float(input("enter the marks of english ="))
math = float(input("enter the marks of math ="))
science = float(input("enter the marks in science ="))
percentage = (english + math + science)/3.0
print("student name is",name,"and the percentage is",percentage)
if percentage >= 90:
    print("grade is A")
elif percentage >= 80:
    print("grade is B")
elif percentage >= 70:
    print("grade is C")
else:
    print ("fail")       
```
